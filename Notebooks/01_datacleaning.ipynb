```python
for col in df.columns:
  print(f"{col}:\n Unique values:",df[col].nunique(),"\n", df[col].unique())
  print()
  print()
```

```python
# Step 1: Select only numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Step 2: Create a dictionary to store outlier counts
outlier_counts = {}

# Step 3: Loop through each numerical column and apply IQR method
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]

# Step 4: Display columns with outliers
for col, count in outlier_counts.items():
    if count > 0:
        print(f"{col}: {count} outliers")
```

```python
# Step 1: Select only numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Step 2: Filter out binary columns (those with only 2 unique values)
continuous_cols = [col for col in numerical_cols if df[col].nunique() > 2]

# Step 3: Create a dictionary to store outlier counts
outlier_counts = {}

# Step 4: Loop through each continuous column and apply IQR method
for col in continuous_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]

# Step 5: Display columns with outliers
print("Outlier counts in continuous columns:")
for col, count in outlier_counts.items():
    if count > 0:
        print(f"{col}: {count} outliers")
```

```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Separate binary and continuous columns
binary_cols = [col for col in numerical_cols if df[col].nunique() == 2]
continuous_cols = [col for col in numerical_cols if df[col].nunique() > 2]
```

```python
# Cap outliers in Elevation and Slope
for col in ['Elevation', 'Slope']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] > upper, upper,
                np.where(df[col] < lower, lower, df[col]))

```

```python
# Drop columns with only one unique value (e.g., Soil_Type_15)
for col in df.columns:
    if df[col].nunique() == 1:
        print(f"Dropping constant column: {col}")
        df.drop(columns=[col], inplace=True)
```

```python
#  Calculate skewness for each continuous column
print("Skewness values:")
for col in continuous_cols:
    skew_val = df[col].skew()
    print(f"{col}: {skew_val:.2f}")
```

```python
#######find skewness
# Apply log1p to skewed distance features
for col in ['Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology']:
    if df[col].min() < 0:
        df[col] = df[col] + abs(df[col].min())  # Shift to positive
    df[col] = np.log1p(df[col])
```

```python
import numpy as np

def transform_skewed_columns(df, threshold=1.0):
    """
    Detects skewed continuous columns and applies log1p transformation
    to reduce right skewness. Only applies to columns with skewness > threshold.
    
    Parameters:
    - df: pandas DataFrame
    - threshold: skewness value above which transformation is applied (default = 1.0)
    
    Returns:
    - df: transformed DataFrame
    """
    
    # Step 1: Select numerical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    
    # Step 2: Filter out binary columns (those with only 2 unique values)
    continuous_cols = [col for col in numerical_cols if df[col].nunique() > 2]
    
    # Step 3: Detect and transform skewed columns
    for col in continuous_cols:
        skew_val = df[col].skew()
        if skew_val > threshold:
            print(f"Transforming {col} (skewness: {skew_val:.2f})")
            
            # Shift if negative values exist
            if df[col].min() < 0:
                df[col] = df[col] + abs(df[col].min())
            
            # Apply log1p transformation
            df[col] = np.log1p(df[col])
    
    return df
```

```python
df = transform_skewed_columns(df)
```
